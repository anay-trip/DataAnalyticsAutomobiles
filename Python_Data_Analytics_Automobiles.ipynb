# Dealing with missing data

1. drop data  
    a. drop the whole row  
    b. drop the whole column  
2. replace data  
    a. replace it by mean  
    b. replace it by frequency  
    c. replace it based on other functions  
    

```python
# Calculate the average of the column
avg_norm_loss = df["normalized-losses"].astype("float").mean(axis=0)
print("Average of the normalized losses: ", avg_norm_loss)

# Replace NaN by mean value in "normalized-losses" column
df["normalized-losses"].replace(np.nan, avg_norm_loss, inplace=True)

# Calculate the mean value for 'bore' column
avg_bore = df['bore'].astype('float').mean(axis=0)
print("Average of bore: ", avg_bore)

# Replace Nan by mean value of "bore" column
df["bore"].replace(np.nan, avg_bore, inplace=True)

# Calculate the mean value of stroke column
avg_stroke = df['stroke'].astype('float').mean(axis=0)
print("Average of stroke: ", avg_stroke)

# Replace NaN by mean value of "stroke" column 
df["stroke"].replace(np.nan, avg_stroke, inplace=True)

# Calculate the mean value of horsepower column and replace NaN values with mean
avg_horsepower = df['horsepower'].astype('float').mean(axis=0)
print("Average horsepower:", avg_horsepower)
df['horsepower'].replace(np.nan, avg_horsepower, inplace=True)

# Calculate the mean value of peak-rpm column and replace NaN values with mean
avg_peakrpm=df['peak-rpm'].astype('float').mean(axis=0)
print("Average peak rpm:", avg_peakrpm)
df['peak-rpm'].replace(np.nan, avg_peakrpm, inplace=True)
```

```python
# To see what values are present in a particular column
df['num-of-doors'].value_counts()

# To calculate the most common type automatically
df['num-of-doors'].value_counts().idxmax()

# To replace the missing number of doors with the most frequent quantity
df['num-of-doors'].replace(np.nan, "four", inplace=True)
```

```python
# To drop whole row with NaN in "price" column
df.dropna(subset=["price"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)

df.head()
```

# ANOVA

### To check if there is statistical difference between two or more groups  

Anova returns two parameters:  
1. F-test score - A larger score means larger difference between the means of the groups
2. P-value - P-value tells us how significant our calculate score is.

```python
grouped_test2 = df_gptest[['drive-wheels', 'price']].groupby(['drive-wheels'])
grouped_test2.head(2)
```

```python
df_gptest

#To obtain values of the method group
grouped_test2.get_group('4wd')['price']
```

```python
#ANOVA
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])
print("ANOVA results: F=", f_val, ", P=", p_val)

# This is a great result, with a large F test score showing a strong correlation and a P value of almost 0 
# implying almost certain statistical significance

#ANOVA seperately for fwd and rwd
f_val, p_val = stats.f_oneway(grouped_test2.get_group('fwd')['price'], grouped_test2.get_group('rwd')['price'])   
print( "ANOVA results: F=", f_val, ", P =", p_val )

#ANOVA seperately for 4wd and rwd
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('rwd')['price'])     
print( "ANOVA results: F=", f_val, ", P =", p_val) 

#ANOVA serperately for 4wd and fwd
f_val, p_val = stats.f_oneway(grouped_test2.get_group('4wd')['price'], grouped_test2.get_group('fwd')['price'])   
print("ANOVA results: F=", f_val, ", P =", p_val)   
```

# Polynomial Regression and Pipelines

```python
# Function to plot the data

def PlotPolly(model, independent_variable, dependent_variable, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)
    
    plt.plot(independent_variable, dependent_variable, '.', x_new, y_new, '-')
    plt.title('Polynomial Fit with Matplotlib for Price ~ Length')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('price of cars')
    
    plt.show()
    plt.close()
```

```python
# to get the variables
x = df['highway-mpg']
y = df['price']

# To fit the polynomial of 3rd order, and to display the polynomial
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)

#To plot the function
PlotPolly(p, x, y, 'highway-mpg')

print("On fitting the polynomial: ", f)
```

```python
# To see if the R^2 values changes over different order polynomials, and plot the results
Rsqu_test = []

order = [1, 2, 3, 4]
for n in order:
    pr = PolynomialFeatures(degree=n)
    
    x_train_pr = pr.fit_transform(x_train[['horsepower']])
    
    x_test_pr = pr.fit_transform(x_test[['horsepower']])    
    
    lr.fit(x_train_pr, y_train)
    
    Rsqu_test.append(lr.score(x_test_pr, y_test))

plt.plot(order, Rsqu_test)
plt.xlabel('order')
plt.ylabel('R^2')
plt.title('R^2 Using Test Data')
plt.text(3, 0.75, 'Maximum R^2 ') 

# We see the R^2 gradually increases until an order three polynomial is used.
# Then the R^2 dramatically decreases at four.
```

# Grid Search

```python
# The term Alfa is a hyperparameter,
# sklearn has the class GridSearchCV to make the process of finding the best hyperparameter simpler.

from sklearn.model_selection import GridSearchCV

# Create a dictionary of parameters
parameters1= [{'alpha': [0.001,0.1,1, 10, 100, 1000, 10000, 100000, 100000]}]

# Create a ridge regions object:
RR=Ridge()

# Create a ridge grid search object:
Grid1 = GridSearchCV(RR, parameters1,cv=4)

# Fit the model
Grid1.fit(x_data[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']], y_data)

# The object finds the best parameter values on the validation data. We can obtain the estimator with the best parameters and assign it to the variable BestRR as follows:
BestRR=Grid1.best_estimator_
print(BestRR)

# To test the model on the test data
BestRR.score(x_test[['horsepower', 'curb-weight', 'engine-size', 'highway-mpg']], y_test)
```
